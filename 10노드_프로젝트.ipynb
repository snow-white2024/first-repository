```python
#자란다 파이썬 10번 프로젝트 [사각형 넓이 구하기]

# 사각형 넓이를 구하는 클래스
class Square:
    def __init__(self):
      while True:
        self.square = int(input('넓이를 구하고 싶은 사각형의 숫자를 써주세요.\n 1.직사각형 2.평행사변형 3.사다리꼴 \n >>>'))

        if self.square == 1:
            print('직사각형 함수는 rect()입니다.')
            print(self.rect())
        elif self.square == 2:
            print('평행사변형 함수는 par()입니다.')
            print(self.par())

        elif self.square == 3:
            print('사다리꼴 함수는 trape()입니다.')
            print(self.trape())
        else:
            print('1, 2, 3 중에서 다시 입력해주세요')
         # 계산이 끝난 후, 다시 계산할 것인지 묻기
        onemore = input("다른 사각형의 넓이를 구하시겠습니까? (y/n): ").strip() .lower()
        if onemore != 'y':
          print("프로그램을 종료합니다.")
          break # 'y'가 아니면 루프를 종료하고 프로그램 종료


    def rect(self):
        while True:
          try:
            width, vertical = map(int, input('가로, 세로를 입력하세요. 예시 : 가로,세로\n >>>').split(','))
            area = width * vertical
            result = '직사각형의 넓이는 : ' + str(area)
            return result

          except ValueError:
            print("숫자를 다시 입력해주세요.")

    def par(self):
      while True:
          try:
            base, high = map(int, input('밑변, 높이를 입력하세요. 예시 : 밑변,높이\n >>>').split(','))  #밑변, 높이를 입력받는 코드
            area = base * high   # 넓이를 구하는 부분
            result = f'평행사변형의 넓이는 : {area}'
            return result

          except ValueError:
            print("숫자를 다시 입력해주세요.")

    def trape(self):
      while True:
          try:
            base, top , high= map(int, input('밑변, 윗변 , 높이를 입력하세요. 예시 : 밑변,윗변,높이\n >>>').split(','))  #밑변, 높이를 입력받는 코드
            area = (base + top) * high/2    # 넓이를 구하는 부분
            result = f'사다리꼴의 넓이는 : {area}'
            return result

          except ValueError:
            print("숫자를 다시 입력해주세요.")

a = Square()
```
